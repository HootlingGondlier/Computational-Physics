```python
import matplotlib.pyplot as plt
import numpy as np
c=3*(10**8)
t1=np.linspace(0,1,100000)

z1=(c/10)*np.sqrt(1+((t1**2)*100))

x=c*t1
plt.style.use('ggplot')


plt.plot(x,z1)

plt.xlabel('ct (m)')
plt.ylabel('z (m)')
plt.title('World line for small launch velocities')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

g=10
u=5
l=10

T = l / u
w0 = 0.5 * g * T
t = np.linspace(0, T)
x = u*t
z1 = (w0*t)-(0.5*g*t**2)

plt.style.use('ggplot')


plt.plot(x,z1)


plt.xlabel('l (m)')
plt.ylabel('z (m)')
plt.title('Spatial Trajectory for Baseball')
plt.show()
```

```python
c=3*10**8
x=c*t
plt.plot(x,z1)
plt.xlabel('ct (m)')
plt.ylabel('z (m)')
plt.title('World line for Baseball')
plt.show()
```
